```python
import matplotlib.pyplot as plt
import numpy as np

def q2_var(step):
    if step<5*600:
        return 40+step*20/3000
    elif step<20*600:
        return 60
    elif step<25*600:
        return 60-(step-20*600)*20/3000
    else:return 40
def q1_var(step):
    if step<5*600:
        return 5+step*5/3000
    elif step<20*600:
        return 10
    elif step<25*600:
        return 10-(step-20*600)*5/3000
    else:return 5
def q1_1(step):
    return 10
def q2_1(step):
    return 60
def q1_2(step):
    return np.random.poisson(10)
def q2_2(step):
    return np.random.poisson(60)


q1=[]
q2=[]
q1_1l=[]
q2_1l=[]
q1_2l=[]
q2_2l=[]
for i in range(18000):
    q1.append(q1_var(i))
    q2.append(q2_var(i))
    q1_1l.append(q1_1(i))
    q2_1l.append(q2_1(i))
    q1_2l.append(q1_2(i))
    q2_2l.append(q2_2(i))

plt.plot(q1)
plt.plot(q2)
```

```python
q1=[]
q2=[]
for i in range(18000):
    q1.append(q1_var(i))
    q2.append(q2_var(i))

plt.figure(figsize=(15, 15))
plt.subplot(3, 1,3)
plt.plot(q1, '-', lw=2,label='q1')  
plt.plot(q2, '--', lw=2,label='q2')          
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
# 设置坐标标签字体大小
plt.xlabel(...,fontsize=18)
plt.ylabel(...,fontsize=18)
plt.xlabel('time step (0.1s)')
plt.ylabel('flow rate (veh/min)')
plt.title('Unimodal demand mode',fontsize=18)
plt.xlim((0,18000))
plt.ylim((0,70))
plt.grid(True)
plt.legend(fontsize=15,loc='upper right')

plt.subplot(3, 1, 1)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.plot(q1_1l, '-', lw=2,label='q1')  
plt.plot(q2_1l, '--', lw=2,label='q2')  
# 设置坐标标签字体大小
plt.xlabel(...,fontsize=18)
plt.ylabel(...,fontsize=18)
plt.xlabel('time step (0.1s)')
plt.ylabel('flow rate (veh/min)')
plt.title('Constant demand',fontsize=18)
plt.xlim((0,18000))
plt.ylim((0,70))
plt.legend()
plt.grid(True)
plt.legend(fontsize=15,loc='upper right')

plt.subplot(3, 1, 2)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.scatter(np.arange(18000),q1_2l,s=1,label='q1')  
plt.scatter(np.arange(18000),q2_2l,s=1,label='q2')  
# 设置坐标标签字体大小
plt.xlabel(...,fontsize=18)
plt.ylabel(...,fontsize=18)
plt.xlabel('time step (0.1s)')
plt.ylabel('flow rate (veh/min)')
plt.title('Constant demand for Poisson distribution',fontsize=18)
plt.xlim((0,18000))
plt.ylim((0,75))
plt.grid(True)
plt.legend(fontsize=15,loc='upper right')#添加图例




plt.tight_layout(pad=0.4, w_pad=0, h_pad=2)
```

```python
import matplotlib.pyplot as plt
# 设置数据
x_values = [10, 15, 20, 25, 30]
y1_values = [0.1,0.4,0.2,0.2,0.1]
y2_values = [0,0.2,0.3,0.3,0.2]


# 绘制柱状图
bar_width =1.5  # 每个柱子的宽度
x1 = [x-bar_width/1.5 for x in x_values]
x3 = [x+bar_width/1.5 for x in x_values]
plt.bar(x1, y1_values, width=bar_width, label='VOT_1',edgecolor=None,color='#00b7ff')
plt.bar(x3, y2_values, width=bar_width, label='VOT_2',edgecolor=None,color='#feb692')
plt.Figure((15,15),dpi=800)
# 添加标题和轴标签
plt.xlabel("VOT($/hr)")
plt.ylabel("VOT probability distribution")
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
# 添加图例
plt.legend()
plt.grid(axis="y")
# 显示图形
plt.show()
```
